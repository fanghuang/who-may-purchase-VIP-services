```python
def get_time(data, var):
    lst = []
    for j in data:
        tmp = [j.month, j.day, j.hour, j.minute]
        lst += [tmp[var]]
    return lst
data_df['month'] = map(lambda x: get_time(x, 0), data_df['activity_time'])
data_df['day'] = map(lambda x: get_time(x, 1), data_df['activity_time'])
data_df['hour'] = map(lambda x: get_time(x, 2), data_df['activity_time'])
data_df['minute'] = map(lambda x: get_time(x, 3), data_df['activity_time'])
```

```python
# basic month, day, hour feature
lst_month = month.keys()
for i in lst_month:
    lst_feature += ['m_'+str(i)]
    data_df['m_'+str(i)] = map(lambda x: 1 if i in x else 0, data_df['month'])
lst_day = day.keys()
for i in lst_day:
    lst_feature += ['d_'+str(i)]
    data_df['d_'+str(i)] = map(lambda x: 1 if i in x else 0, data_df['day'])
lst_hour = hour.keys()
for i in lst_hour:
    lst_feature += ['h_'+str(i)]
    data_df['h_'+str(i)] = map(lambda x: 1 if i in x else 0, data_df['hour'])

```

```python
# check if time in a specific range
def check_range(lst, start, end):
    for i in lst:
        if i in range(start, end):
            return 1
    return 0

# extra month feature separate month to 3 bucket
data_df['m_4_7'] = map(lambda x: check_range(x, 4, 7), data_df['month'])
data_df['m_7_10'] = map(lambda x: check_range(x, 7, 10), data_df['month'])
data_df['m_10_12'] = map(lambda x: check_range(x, 10, 13), data_df['month'])

# extra day feature separate month to 3 bucket
data_df['d_1_11'] = map(lambda x: check_range(x, 1, 11), data_df['day'])
data_df['d_11_21'] = map(lambda x: check_range(x, 11, 21), data_df['day'])
data_df['d_21_32'] = map(lambda x: check_range(x, 21, 32), data_df['day'])

# extra hour feature separate month to 3 bucket
data_df['h_0_6'] = map(lambda x: check_range(x, 0, 6), data_df['hour'])
data_df['h_6_12'] = map(lambda x: check_range(x, 6, 12), data_df['hour'])
data_df['h_12_18'] = map(lambda x: check_range(x, 12, 18), data_df['hour'])
data_df['h_18_24'] = map(lambda x: check_range(x, 18, 24), data_df['hour'])

lst_feature += ['m_4_7','m_7_10','m_10_12','d_1_11','d_11_21','d_21_32',
                'h_0_6','h_6_12','h_12_18','h_18_24']
```

```python
import scipy as sp
def logloss(act, pred):
    epsilon = 1e-15
    pred = sp.maximum(epsilon, pred)
    pred = sp.minimum(1-epsilon, pred)
    ll = sum(act*sp.log(pred) + sp.subtract(1,act)*sp.log(sp.subtract(1,pred)))
    ll = ll * -1.0/len(act)
    return ll
```
